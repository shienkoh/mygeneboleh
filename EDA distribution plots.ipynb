```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("full_data.csv")
```

```python
data.head()
```

```python
fig, ax = plt.subplots(1, 1)
sns.kdeplot(data["Var of -1 Means"],label='-1',ax=ax)
sns.kdeplot(data["Var of 0 Means"],label='0',ax=ax)
sns.kdeplot(data["Var of +1 Means"],label='+1',ax=ax)
ax.set_xlim(0, 60)
ax.set(xlabel='Variance of Means', ylabel='Density')
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots(1, 1)
sns.kdeplot(data["Var of -1 Dwell"],label='-1',ax=ax)
sns.kdeplot(data["Var of 0 Dwell"],label='0',ax=ax)
sns.kdeplot(data["Var of +1 Dwell"],label='+1',ax=ax)
ax.set_xlim(0, 0.0001)
ax.set(xlabel='Variance of Dwell', ylabel='Density')
ax.legend()
plt.locator_params(nbins=5)
plt.show()
```

```python
fig, ax = plt.subplots(1, 1)
sns.ecdfplot(data["Var of -1 SD"],label='-1',ax=ax)
sns.ecdfplot(data["Var of 0 SD"],label='0',ax=ax)
sns.ecdfplot(data["Var of +1 SD"],label='+1',ax=ax)
ax.set(xlabel='Variance of SD', ylabel='Density')
ax.set_xlim(0, 20)
ax.legend()
plt.show()
```
