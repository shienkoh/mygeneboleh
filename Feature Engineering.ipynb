```python
import pandas as pd
import numpy as np
import statistics 
import math
from ast import literal_eval
```

```python
data = pd.read_csv("merged.csv")
```

```python
data.head(10)
```

```python
data['Reads'] = data['Reads'].apply(literal_eval)
```

```python
read_lengths = []
for i in range(len(data)):
    read_lengths.append(len(data['Reads'][i]))
```

```python
data['Read Lengths'] = read_lengths
```

```python
data.head(10)
```

Seperating each of the 3 x 3 features into individual columns



```python
dwell_left_final = []
dwell_mid_final = []
dwell_right_final = []
mean_left_final = []
mean_right_final = []
sd_left_final = []
sd_mid_final = []
sd_right_final = []

for i in range(len(data)):
    dwell_left = []
    dwell_mid = []
    dwell_right = []
    mean_left = []
    mean_right = []
    sd_left = []
    sd_mid = []
    sd_right = []
    for read in data['Reads'][i]:
        dwell_left.append(read[0])
        sd_left.append(read[1])
        mean_left.append(read[2])
        
        dwell_mid.append(read[3])
        sd_mid.append(read[4])
        
        dwell_right.append(read[6])
        sd_right.append(read[7])
        mean_right.append(read[8])
        
    dwell_left_final.append(dwell_left)
    dwell_mid_final.append(dwell_mid)
    dwell_right_final.append(dwell_right)
    
    sd_left_final.append(sd_left)
    sd_mid_final.append(sd_mid)
    sd_right_final.append(sd_right)
    
    mean_left_final.append(mean_left)
    mean_right_final.append(mean_right)
```

```python
data['Signal Dwell Time -1'] = dwell_left_final
data['Signal Dwell Time 0'] = dwell_mid_final
data['Signal Dwell Time +1'] = dwell_right_final

data['Means -1'] = mean_left_final
data['Means +1'] = mean_right_final

data['SD -1'] = sd_left_final
data['SD 0'] = sd_mid_final
data['SD +1'] = sd_right_final
```

m6A occurs at DRACH motifs :

D = A,G or U
R = A or G
H = A, C or T


So the String sequence is of the format : H-DRACH-D

Since 3rd and 4th position (index 3 and 4) are guaranteed A and C, don't need to use as a feature

```python
left = []
right = []
first = []
second = []
fifth = []
for i in range(len(data)):
    five_mer = data['5-mers'][i]
    
    #for D, A = 1, G = 2, U = 3
    
    if five_mer[1] == 'A' : 
        first.append(1)
    elif five_mer[1] == 'G' :
        first.append(2)
    else :
        first.append(3)
        
    if five_mer[6] == 'A' : 
        right.append(1)
    elif five_mer[6] == 'G' :
        right.append(2)
    else :
        right.append(6)
    
    #for R, A = 1, G = 2
    
    if five_mer[2] == 'A' :
        second.append(1)
    else :
        second.append(2)
    
    #for H, A = 1, C = 2, T = 3
    
    if five_mer[5] == 'A':
        fifth.append(1)
    elif five_mer[5] == 'C':
        fifth.append(2)
    else :
        fifth.append(3)
        
    if five_mer[0] == 'A':
        left.append(1)
    elif five_mer[0] == 'C':
        left.append(2)
    else :
        left.append(3)
```

```python
data['right 5-mer'] = left
data['1-mer'] = first
data['2-mer'] = second
data['5-mer'] = fifth
data['right 1-mer'] = right
```

```python
data.head(10)
```

```python
data['Means'] = data['Means'].apply(literal_eval)
```

```python
mean_of_left_means = []
mean_of_right_means = []
mean_of_mid_means = []
mean_of_left_sd = []
mean_of_mid_sd = []
mean_of_right_sd = []
mean_of_left_dwell= []
mean_of_mid_dwell= []
mean_of_right_dwell= []

for i in range(len(data)):
    mean_of_left_means.append(statistics.mean(data['Means -1'][i]))
    mean_of_mid_means.append(statistics.mean(data['Means'][i]))
    mean_of_right_means.append(statistics.mean(data['Means +1'][i]))
    mean_of_left_sd.append(statistics.mean(data['SD -1'][i]))
    mean_of_mid_sd.append(statistics.mean(data['SD 0'][i]))
    mean_of_right_sd.append(statistics.mean(data['SD +1'][i]))
    mean_of_left_dwell.append(statistics.mean(data['Signal Dwell Time -1'][i]))
    mean_of_mid_dwell.append(statistics.mean(data['Signal Dwell Time 0'][i]))
    mean_of_right_dwell.append(statistics.mean(data['Signal Dwell Time +1'][i]))
```

```python
data['Mean of -1 Means'] = mean_of_left_means
data['Mean of 0 Means'] = mean_of_mid_means
data['Mean of +1 Means'] = mean_of_right_means
data['Mean of -1 SD'] = mean_of_left_sd
data['Mean of 0 SD'] = mean_of_mid_sd
data['Mean of +1 SD'] = mean_of_right_sd
data['Mean of -1 Dwell'] = mean_of_left_dwell
data['Mean of 0 Dwell'] = mean_of_mid_dwell
data['Mean of +1 Dwell'] = mean_of_right_dwell
```
